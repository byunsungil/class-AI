하나의 함수를 만들어 0~12미만 아침, 12~18미만 점심, 18~24 저녁

```python
def x(a):
    if a < 12:
        print("아침")
    elif a < 18:
        print("점심")
    elif a < 24:
        print("저녁")
```

```python
x(23)
```
